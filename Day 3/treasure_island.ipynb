```python
print("Welcome to Treasure Island.")
print("Your mission is to find the treasure.")

choice1 = input("enter the direction left or right?\n").lower()
if choice1 == "left":
  choice2 = input("You\'ve come to a lake. There is an island in the middle of the lake. Type 'wait' to wait for a boat. Type 'swim' to swim across.\n").lower()
  if choice2 == "wait":
    choice3 = input("You arrived at the island unharmed. There is a house with 3 doors. One red, one blue and one yellow. Which colour do you choose?\n").lower()
    if choice3 == "red":
      print("Game Over ")
    elif choice3 == "blue":
      print("Game Over ")
    else:
      print("You won!")
  else:
    print("Game Over ")
else:
  print("Game Over ")
```
